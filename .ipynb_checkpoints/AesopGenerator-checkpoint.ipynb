```python
#read data into dataframe
with open('aesopFables.json') as f:
    data = json.load(f)
    jsonData = data["stories"]
    data = pd.DataFrame(list(jsonData))#, columns =['title', 'story', 'moral']
```

```python
fable = [x.lower() for x in data["story"][1]]
fable = [re.sub('[^a-zA-Z]', ' ', x) for x in fable]
print(fable)
```

```python
#Tokenize all fables
all_fables = []
for story in data["story"]:
    
    curr_fable = [x.lower() for x in story]
    curr_fable = [re.sub('([.,!?()])', r' \1 ', x) for x in curr_fable]
    all_fables.append(curr_fable)
#print(all_fables)
```

```python
test_text = nltk.word_tokenize(all_fables[0][0])
tagged_tokens = nltk.pos_tag(test_text)
#print(tagged_tokens)

grammar_dict = {}
pos_fables = []

for curr_fable in all_fables:
    pos_list = []
    for sentence in curr_fable:
        token_sent = nltk.word_tokenize(sentence)
        tagged_sent = nltk.pos_tag(token_sent)
        pos_list.append(tagged_sent)
        for pair in tagged_sent:
            if pair[1] in grammar_dict:
                #grammar_dict[pair[1]] += set((pair[0]))
                #rint(grammar_dict[pair[1]])
                #print(pair[0])
                grammar_dict[pair[1]].add(pair[0])
            else:
                grammar_dict[pair[1]] = set([pair[0]])
    pos_fables.append(pos_list)
    #print ("Fable" + str(fable_num))
    #print(curr_fable)
            
print(pos_fables)
```

```python
#Let's take a look at the general grammatical structure
pos_simplified = [] 
for curr_fable in pos_fables:
    sent_tags = []
    for curr_sentence in curr_fable:
        #pos_sent = ""
        pos_simp = []
        for curr_token in curr_sentence:
            #pos_sent += curr_token[1] + " "
            pos_simp.append(curr_token[1])
        #print(pos_sent)
        sent_tags.append(pos_simp)
    pos_simplified.append(sent_tags)
print(pos_simplified)
```

```python
# This first step takes crunches down all vebrs, nouns, etc. into noun phrases, etc. as it goes
# Input is a list of tagged POS documents, where each POS documents is a the fable converted to POS tag
def apply_ngram_step(corpus, grammar, replacement):
    #Applies a grammar to a corpus. Used as a middle step in the layer simplifications
    str_gramm = [" ".join(x) for x in grammar]
    new_corpus = []
    for curr_sent in corpus:
        new_sent = curr_sent
        for rule in str_gramm:
            new_sent = re.sub(r"\b{}\b".format(rule), replacement, new_sent)
            #new_sent = new_sent.replace(rule, replacement)
        new_corpus.append(new_sent)
    return new_corpus

def sents_tokenization(corpus):
    sent_toks = [WordPunctTokenizer().tokenize(x) for x in corpus]
    return sent_toks

#print(apply_ngram_step(set_pos, test_gram, "CONJ"))    

def first_layer_simplification(corpus, vocab):
    #Apply phrase conversions in reverse order to the heirarchy graph? Coordinations to NP order. Conjunctions/Prepositions are combined in trigrams
    bigram_measures = nltk.collocations.BigramAssocMeasures()
    trigram_measures = nltk.collocations.TrigramAssocMeasures()
    
    punc = [',', '.']
    conjs = [':', 'CC']
    prepp = ['IN']
    np = ['NN', 'NNP', 'NNPS', 'NNS', 'PRP']
    vp = ['VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ']
    
    #the first step is working on np, then vp, then prepp, then conjs
    
    np_tokens = sents_tokenization(corpus)
    finder1 = BigramCollocationFinder.from_documents(np_tokens)
    finder1.apply_word_filter(lambda w: w in vp+prepp+conjs+punc)
    finder1.apply_ngram_filter((lambda w1, w2: all(y not in (w1, w2) for y in np )))
    np_gram = sorted(finder1.above_score(bigram_measures.raw_freq, 0))
    
    corp2 = apply_ngram_step(corpus, np_gram, "NP")
    #then, we feed the new corpus back in for vps, prepp, etc.
    #print(corp2)
    
    vp_tokens = sents_tokenization(corp2)
    finder2 = BigramCollocationFinder.from_documents(vp_tokens)
    finder2.apply_word_filter(lambda w: w in prepp+conjs+punc)
    finder2.apply_ngram_filter((lambda w1, w2: all(y not in (w1, w2) for y in vp )))
    vp_gram = sorted(finder2.above_score(bigram_measures.raw_freq, 0)) # this controls the filtering at the moment
    
    corp3 = apply_ngram_step(corp2, vp_gram, "VP")
    #print(corp3)
    
    prepp_tokens = sents_tokenization(corp3)
    finder3 = TrigramCollocationFinder.from_documents(prepp_tokens)
    finder3.apply_word_filter(lambda w: w in conjs+punc)
    finder3.apply_ngram_filter(lambda w1, w2, w3: all(y not in (w2) for y in prepp))
    prepp_gram = sorted(finder3.above_score(bigram_measures.raw_freq, 0)) # this controls the filtering at the moment
    
    corp4 = apply_ngram_step(corp3, prepp_gram, "PREPP ")
    #print(corp4)
    
    conj_tokens = sents_tokenization(corp4)
    finder4 = TrigramCollocationFinder.from_documents(conj_tokens)
    finder4.apply_word_filter(lambda w: w in punc)
    finder4.apply_ngram_filter(lambda w1, w2, w3: all(y not in (w2) for y in conjs))
    conj_gram = sorted(finder4.above_score(bigram_measures.raw_freq, 0))
    
    corp5 = apply_ngram_step(corp4, conj_gram, "CONJP ")
    #print(corp5)
    
    #For the conj and prepp steps, consider also doing another step of simplification, this time for bigrams instead of trigrams.
    
    new_grammar = vocab
    new_grammar["NP"] = np_gram
    new_grammar["VP"] = vp_gram
    new_grammar["PREPP"] = prepp_gram
    new_grammar["CONJP"] = conj_gram
    
    # Then we combine all the grammars together as part of the grammar dict. Perhaps we filter for the top n-grams in the final corpus to create the source?
    return new_grammar, corp5
gen_gram, fin_corp = first_layer_simplification(set_pos, grammar_dict)
#run a bigram top ten filter for NP, VP, CP, etc. Form rules based off of that. Conduct first grammar validity test.
```

```python
#parsing the final grammar dictionary into a grammar string for nltk
def grammar_string(gram):
    sgram = ""
    for key in gram:
        val = ""
        if type(gram[key]) is tuple:
            print(gram[key])
        else:
            for v in gram[key]:
                if type(v) is not tuple:
                    val+=(" " + v + " |")
                else:
                    for i in v:
                        val+=(i + " ")
                    val+="| "
            val = val[1:-2]
            sgram+=(key + " -> " + val + "\n")
    print(sgram)
            
grammar_string(gen_gram)
```

```python
# CAN DELETE BLOCK

tokenizer = Tokenizer()
tokenizer.fit_on_texts(fable) #creates tokens of each words as a dictionary with key being word and the value beings its token
total_words = len(tokenizer.word_index) + 1 #calculating total number of words in the initial sentence

print(tokenizer.word_index)
print(total_words)
```

```python
# CAN DELETE BLOCK

input_sequences = [] #training features (xs) will be a list

for line in fable:
    token_list = tokenizer.texts_to_sequences([line])[0] #converts each sentence as its tokenized equivalent
    for i in range(1, len(token_list)):
        n_gram_sequence = token_list[:i+1] #generating n gram sequences: 1st two words in the token_list is one sequence, next two words are another sequence and so on
        input_sequences.append(n_gram_sequence) #appending each n gram sequence to the list of our features (xs)

# pad sequences 
max_sequence_len = max([len(x) for x in input_sequences]) #calculating the length of the longest sequence in our training features (xs)
input_sequences = np.array(pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre')) #pre-pading each value of the input_sequence
# create predictors and label
xs, labels = input_sequences[:,:-1],input_sequences[:,-1] #creating xs and their labels using numpy slicing
ys = tf.keras.utils.to_categorical(labels, num_classes=total_words) #creating one hot encoding values of each labels to make our ys
```

```python
# CAN DELETE BLOCK

model = Sequential() #creating a sequential model
model.add(Embedding(total_words, 64, input_length=max_sequence_len-1)) #adding an embedding layer with 64 as the embedding dimension
model.add(Bidirectional(LSTM(20))) #adding 20 LSTM units
model.add(Dense(total_words, activation='softmax')) #creating a dense layer with 54 output units (total_words) with softmax activation
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy']) #compiling the model with adam optimiser
history = model.fit(xs, ys, epochs=500, verbose=1) #training for 500 epochs
```

```python
# CAN DELETE BLOCK

#predicting the next word using an initial sentence
seed_text = "the tortoise carries the house"
next_words = 10
  
for _ in range(next_words):
    token_list = tokenizer.texts_to_sequences([seed_text])[0] #converting our seed_text to tokens and excluding the out of vcabulary words
    token_list = pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre') #padding the seed_text
    predicted = model.predict(token_list, verbose=0) #predicting the token of the next word using our trained model
    predicted = np.argmax(predicted,axis=1)
    output_word = "" #initialising output word as blank at the beginning
    for word, index in tokenizer.word_index.items():
        if index == predicted:
            output_word = word #converting the token back to the corresponding word and storing it in the output_word
            break
    seed_text += " " + output_word
print(seed_text)
```
